```python
from math import *
print("ax**2+b*x+c=0")
a=float(input("Nhập vào a: "))
b=float(input("Nhập vào b: "))
c=float(input("Nhập vào c: "))
print(f"{a}x**2+{b}*x+{c}=0")
delta=b**2-4*a*c
if delta>0:
  print(f"Chương trình có 2 nghiệm: x1={(-b+sqrt(delta))/(2*a)} và \n x2={(-b-sqrt(delta))/(2*a)}")
elif delta==0:
  print(f"Chương trình có 1 nghiệm: x={-b/a}")
else:
  print("Chương trình vô nghiệm")
```
